```python
#metodo para criar uma tabela dinamica correlacionando campos por aceitação da ultima campanha

def TabDimEducationXMarital(paramsTxt, paramsValues, operation,flagAccepted):
    x = df[(df["Response"])==flagAccepted].groupby(paramsTxt,as_index=False).agg({paramsValues:operation})
    tab = x.pivot(index=paramsTxt[0],columns=paramsTxt[1],values=paramsValues).fillna(0)
    tab["TotalBase"] = df.value_counts("Education")
    tab["Total Geral"] = tab.sum(axis=1,skipna=True)
    tab.loc['Total'] =  tab.sum(axis=0,skipna=True)
    return tab
    
```

```python

x = df[(df["Response"])==1].groupby(["Education","Marital_Status"],as_index=False).agg({"Response":"count"})
tab = x.pivot(index='Education',columns='Marital_Status',values='Response').fillna(0)
tab["Total Geral"] = tab.sum(axis=1,skipna=True)

tab.loc['Total'] =  tab.sum(axis=0,skipna=True)
tab
```

<b>Cruzando dados de nivel de educação com status de relacionamento, 
vemos que pessoas solteiras e casadas com grau de escolaridade de graduação ou superior aceitaram mais
campanhas</b>

```python
x = df[(df["Response"])==0].groupby(["Education","Marital_Status"],as_index=False).agg({"Response":"count"})
tab = x.pivot(index='Education',columns='Marital_Status',values='Response').fillna(0)
tab["Total Geral"] = tab.sum(axis=1,skipna=True)

tab.loc['Total'] =  tab.sum(axis=0,skipna=True)
tab

```
